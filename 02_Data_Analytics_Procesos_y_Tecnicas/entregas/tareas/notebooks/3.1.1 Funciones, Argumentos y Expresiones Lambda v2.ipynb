Generalmente utilizamos el método <b>return</b> para retornar el valor de la función. Podemos crear una función que escriba la serie de Fibonacci hasta un límite determinado:

```python
def fib(n):  # escribe la serie de Fibonacci hasta n
    """Escribe la serie de Fibonacci hasta n."""
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    print()

# Ahora llamamos a la funcion que acabamos de definir:
fib(200)
```

Es simple escribir una función que retorne una lista con los números de la serie de Fibonacci en lugar de imprimirlos:

```python
def fib2(n): # devuelve la serie de Fibonacci hasta n
    """Devuelve una lista conteniendo la serie de Fibonacci hasta n."""
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)    # ver abajo
        a, b = b, a+b
    return result

f100 = fib2(100)    # llamarla
f100                # escribir el resultado
```

En a clase pasada, hicimos un trecho de código que calculaba una secuencia de ADN. Ahora lo colocaremos en una función:

```python
def reverso_complementar(sequencia):
    complemento=""
    for x in sequencia:
        if x=="G":
            complemento = complemento + "C"
        elif x=="C":
            complemento = complemento + "G"
        elif x == "A":
            complemento = complemento + "T"
        elif x == "T":
            complemento = complemento + "A"
    revcomp=''.join(reversed(complemento))
    return revcomp
```

```python
#Secuencias de ADN son strings sobre el alfabeto {A,C,G,T}
dna = "ACAGCAGAGTGACCCAGT"
print(reverso_complementar(dna))
```

```python
#Una vez definida una función, podemos llamarla varias veces
print(reverso_complementar("ACGCAGGTA"))
```

```python
# ========== FUNCION HEADER ==========
def header(texto):
    mult = len(texto) + 8
    print("╬" + "═" * mult + "╬")
    print(" " + " " * 4 + texto + " " * 4 + " ")
    print("╬" + "═" * mult + "╬")
    return mult


# ========== FUNCION MOSTRAR LISTA ==========
def mostrar_lista(agenda):
    header("MOSTRAR AGENDA")
    line=50

    if not agenda:
        print("\tAgenda vacia\t")
        print("+" + "═" * line + "+")
        return

    for i, c in enumerate(agenda):
        print(f"\tRegistro:\t\t{i+1}")
        print(f"\tNombre:\t\t{c['nombre']:<38}")
        print(f"\tEmail:\t\t{c['email']:<38}")
        print(f"\tTelefono:\t\t{c['telefono']:<36}")
        print("+" + "-" * line + "+")


# --- EJECUCION ---
mostrar_lista(agenda)
```

### Ejercicio 3:

1.1) Defina una función llamada imc que reciba como argumentos altura (en metros) y peso (en Kg) y retorne el IMC (Índice de Masa Corporal). IMC = peso / altura² <BR>
1.2) Defina otra función que reciba el imc calculado con la função arriba y retorne "delgado","normal","sobrepeso","obeso" conforme el IMC<BR>
delgado: <20  <BR>
normal: >=20 e <25  <BR>
sobrepeso: >=25 e <30  <BR>
obeso: >=30  <BR>

```python
# Escriba el código aqui
# ========== FUNCION 1: CALCULAR IMC ==========
def imc(peso, altura):
    """
    Calcula el Indice de Masa Corporal
    IMC = peso / altura²
    """
    try:
        peso = float(peso)
        altura = float(altura)
        resultado = peso / (altura ** 2)
        return round(resultado, 2)
    except:
        return None


# ========== FUNCION 2: CLASIFICAR IMC ==========
def clasificar_imc(valor_imc):
    """
    Clasifica el IMC segun los rangos:
    delgado: <20
    normal: >=20 y <25
    sobrepeso: >=25 y <30
    obeso: >=30
    """
    if valor_imc is None:
        return "Error"

    if valor_imc < 20:
        return "delgado"
    elif valor_imc < 25:
        return "normal"
    elif valor_imc < 30:
        return "sobrepeso"
    else:
        return "obeso"


# ========== EJECUCION ==========
# Prueba
peso = 70
altura = 1.75

resultado_imc = imc(peso, altura)
clasificacion = clasificar_imc(resultado_imc)

print(f"IMC: {resultado_imc}")
print(f"Clasificacion: {clasificacion}")
```

# 5.2 Más sobre definición de funciones

También es posible definir funciones con un número variable de argumentos. Hay tres formas que pueden ser combinadas.

## 5.2.1 Argumentos con valores por omisión

La forma más útil es especificar un valor por omisión para uno o más argumentos. Esto crea una función que puede ser llamada con menos argumentos que los que permite. Por ejemplo:

```python
def pedir_confirmacion(prompt, reintentos=4, recordatorio='Por favor, intente nuevamente!'):
    while True:
        ok = input(prompt)
        if ok in ('s', 'S', 'si', 'Si', 'SI'):
            return True
        if ok in ('n', 'no', 'No', 'NO'):
            return False
        reintentos = reintentos - 1
        if reintentos < 0:
            raise ValueError('Respuesta de usuario es inválida')
        print(recordatorio)
```

Esta función puede ser llamada de distintas maneras:<br>
<ul>
    <li>pasando sólo el argumento obligatorio: pedir_confirmacion('¿Realmente queres salir?')</li>
    <li>pasando uno de los argumentos opcionales: pedir_confirmacion('¿Sobreescribir archivo?', 2)</li>
    <li>o pasando todos los argumentos: pedir_confirmacion('¿Sobreescribir archivo?', 2, "Vamos, solo si o no!")</li>
</ul>

```python
pedir_confirmacion('¿Realmente queres salir?')
```

```python
pedir_confirmacion('¿Sobreescribir archivo?', 2)
```

```python
pedir_confirmacion('¿Sobreescribir archivo?', 2, "Vamos, solo si o no!")
```

Los valores por omisión son evaluados en el momento de la definición de la función, en el ámbito de la definición, entonces:

```python
i = 5

def f(arg=i):
    print(arg)

i = 6
f()
```

<b>Advertencia importante:</b> El valor por omisión es evaluado solo una vez. Existe una diferencia cuando el valor por omisión es un objeto mutable como una lista, diccionario, o instancia de la mayoría de las clases. Por ejemplo, la siguiente función acumula los argumentos que se le pasan en subsiguientes llamadas:

```python
def f(a, L=[]):
    L.append(a)
    return L

print(f(1))
print(f(2))
print(f(3))
```

Si no se quiere que el valor por omisión sea compartido entre subsiguientes llamadas, se pueden escribir la función así:

```python
def f(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L

print(f(1))
print(f(2))
print(f(3))
```

## 5.2.2 Palabras claves como argumentos

Las funciones también puede ser llamadas usando argumentos de palabras clave (o argumentos nombrados) de la forma keyword = value. Por ejemplo, la siguiente función:

...acepta un argumento obligatorio (tension) y tres argumentos opcionales (estado, accion, y tipo). Esta función puede llamarse de cualquiera de las siguientes maneras:

```python
def loro(tension, estado='muerto', accion='explotar', tipo='Azul Nordico'):
    print("-- Este loro no va a", accion, end=' ')
    print("si le aplicás", tension, "voltios.")
    print("-- Gran plumaje tiene el", tipo)
    print("-- Está", estado, "!")
```

```python
loro(1000)                                          # 1 argumento posicional
```

```python
loro(tension=1000)                                  # 1 argumento nombrado
```

```python
loro(tension=1000000, accion='VOOOOOM')             # 2 argumentos nombrados
```

```python
loro(accion='VOOOOOM', tension=1000000)             # 2 argumentos nombrados
```

```python
loro('un millón', 'despojado de vida', 'saltar')    # 3 args posicionales
```

```python
loro('mil', estado='viendo crecer las flores desde abajo')  # uno y uno
```

...pero estas otras llamadas serían todas inválidas:

```python
loro()                      # falta argumento obligatorio
loro(tension=5.0, 'muerto') # argumento posicional luego de uno nombrado
loro(110, tension=220)      # valor duplicado para el mismo argumento
loro(actor='Juan Garau')    # nombre del argumento desconocido
```

En una llamada a una función, los argumentos nombrados deben seguir a los argumentos posicionales. Cada uno de los argumentos nombrados pasados deben coincidir con un argumento aceptado por la función (por ejemplo, actor no es un argumento válido para la función loro), y el orden de los mismos no es importante. Esto también se aplica a los argumentos obligatorios (por ejemplo, loro(tension=1000) también es válido). Ningún argumento puede recibir más de un valor al mismo tiempo. Aquí hay un ejemplo que falla debido a esta restricción:

```python
def funcion(a):
    pass

funcion(0, a=0)
```

Cuando un parámetro formal de la forma **nombre está presente al final, recibe un diccionario (ver Tipos integrados) conteniendo todos los argumentos nombrados excepto aquellos correspondientes a un parámetro formal. Esto puede ser combinado con un parámetro formal de la forma *nombre (descripto en la siguiente sección) que recibe una tupla conteniendo los argumentos posicionales además de la lista de parámetros formales. (*nombre debe ocurrir antes de **nombre). Por ejemplo, si definimos una función así:

```python
def ventadequeso(tipo, *argumentos, **palabrasclaves):
    print("-- ¿Tiene", tipo, "?")
    print("-- Lo siento, nos quedamos sin", tipo)
    for arg in argumentos:
        print(arg)
    print("-" * 40)
    for c in palabrasclaves:
        print(c, ":", palabrasclaves[c])
```

```python
ventadequeso("Limburger", "Es muy liquido, sr.",
             "Realmente es muy muy liquido, sr.",
             cliente="Juan Garau",
             vendedor="Miguel Paez",
             puesto="Venta de Queso Argentino")
```

Se debe notar que el orden en el cual los argumentos nombrados son impresos está garantizado para coincidir con el orden en el cual fueron provistos en la llamada a la función.

```python
def suma(*numeros):
    s = 0
    for i in numeros:
        s += i
    print(f"La suma de numeros es igual a {s}")
```

```python
suma(1,7,9,34)
```

## 5.2.3 Listas de argumentos arbitrarios

Finalmente, la opción menos frecuentemente usada es especificar que una función puede ser llamada con un número arbitrario de argumentos. Estos argumentos serán organizados en una tupla. Antes del número variable de argumentos, cero o más argumentos normales pueden estar presentes:

```python
def muchos_items(archivo, separador="*", *args):
    archivo.write(separador.join(args))
```

```python
with open("mi_archivo.txt", "w") as archivo:
    muchos_items(archivo, ",", "manzana", "naranja", "plátano", "uva")
```

Normalmente estos argumentos de cantidad variables son los últimos en la lista de parámetros formales, porque toman todo el remanente de argumentos que se pasan a la función. Cualquier parámetro que suceda luego del *args será ‘sólo nombrado’, o sea que sólo se pueden usar como nombrados y no posicionales.:

```python
def concatenar(*args, sep="/"):
    return sep.join(args)
```

```python
concatenar("tierra", "marte", "venus")
```

```python
concatenar("tierra", "marte", "venus", sep="*")
```

## 5.2.4 Desempaquetando una lista de argumentos

La situación inversa ocurre cuando los argumentos ya están en una lista o tupla pero necesitan ser desempaquetados para llamar a una función que requiere argumentos posicionales separados. Por ejemplo, la función predefinida range() espera los argumentos inicio y fin. Si no están disponibles en forma separada, se puede escribir la llamada a la función con el operador para desempaquetar argumentos de una lista o una tupla *::

```python
list(range(3, 6))   # llamada normal con argumentos separados
```

```python
args = [3, 6]
list(range(*args))  # llamada con argumentos desempaquetados de la lista
```

Del mismo modo, los diccionarios pueden entregar argumentos nombrados con el operador **:

```python
def loro(tension, estado='rostizado', accion='explotar'):
    print("-- Este loro no va a", accion, end=' ')
    print("si le aplicás", tension, "voltios.", end=' ')
    print("Está", estado, "!")

d = {'tension':'cinco mil','estado':'demacrado','accion':'VOLAR'}

loro(**d)
```

### Ejercicio 5:

Escriba un programa que pida dos años y escriba cuántos años bisiestos hay entre esas dos fechas (incluidos los dos años):<br>

<b>Entrada:</b><br>
CONTADOR DE AÑOS BISIESTOS<br>
Escriba un año: 2000<br>
Escriba otro año posterior a 2000: 1999<br>

<b>Salida:</b><br>
1999 no es mayor que 2000. Inténtelo de nuevo: 2013<br>

<b>Otra salida:</b><br>
De 2000 a 2013 hay 14 años, 4 de ellos bisiestos.<br>

```python
# Escriba el código aqui
# ========== FUNCION PARA VERIFICAR AÑO BISIESTO ==========
def es_bisiesto(anio):
    """
    Verifica si un año es bisiesto
    Un año es bisiesto si es divisible por 4,
    excepto si es divisible por 100 pero no por 400
    """
    if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
        return True
    return False


# ========== PROGRAMA PRINCIPAL ==========
header("CONTADOR DE AÑOS BISIESTOS")

# Solicitar primer año
while True:
    try:
        anio1 = int(input("Escriba un año: "))
        break
    except ValueError:
        print("\t Error: Ingrese un año valido")

# Solicitar segundo año (debe ser mayor)
while True:
    try:
        anio2 = int(input("Escriba otro año posterior a {}: ".format(anio1)))

        if anio2 > anio1:
            break
        else:
            print(f"\t {anio2} no es mayor que {anio1}. Inténtelo de nuevo:")
    except ValueError:
        print("\t Error: Ingrese un año valido")

# Contar años bisiestos
contador = 0
for anio in range(anio1, anio2 + 1):
    if es_bisiesto(anio):
        contador += 1

# Calcular total de años
total_años = anio2 - anio1 + 1

# Mostrar resultado
print(f"\t De {anio1} a {anio2} hay {total_años} años, {contador} de ellos bisiestos.")
print("-" * 50 + "\n")
```

```python
# Escriba el código aqui
# ========== FUNCION 1: DECIMAL A BINARIO ==========
def decimal_a_binario(decimal):
    """
    Convierte un numero decimal a binario

    Parametros:
    - decimal: numero entero en base 10

    Retorna:
    - string con el numero en binario
    """
    try:
        decimal = int(decimal)

        if decimal < 0:
            return "Error: No se permiten numeros negativos"

        if decimal == 0:
            return "0"

        binario = ""
        num = decimal

        while num > 0:
            binario = str(num % 2) + binario
            num = num // 2

        return binario

    except ValueError:
        return "Error: Ingrese un numero valido"


# ========== FUNCION 2: BINARIO A DECIMAL ==========
def binario_a_decimal(binario):
    """
    Convierte un numero binario a decimal

    Parametros:
    - binario: string con numeros 0 y 1

    Retorna:
    - numero entero en base 10
    """
    try:
        binario = str(binario).strip()

        # Verificar que solo contenga 0 y 1
        for digito in binario:
            if digito not in '01':
                return "Error: El numero binario solo debe contener 0 y 1"

        decimal = 0
        potencia = 0

        for digito in reversed(binario):
            decimal += int(digito) * (2 ** potencia)
            potencia += 1

        return decimal

    except:
        return "Error: Ingrese un binario valido"


# ========== PROGRAMA DE PRUEBA ==========
header("CONVERSOR DECIMAL - BINARIO")

# Prueba 1: Decimal a Binario
print("\n[ DECIMAL A BINARIO ]")
numero = input("Ingrese numero decimal: ")
resultado = decimal_a_binario(numero)
print(f"  {numero} en binario es: {resultado}")

# Prueba 2: Binario a Decimal
print("\n[ BINARIO A DECIMAL ]")
binario = input("Ingrese numero binario: ")
resultado = binario_a_decimal(binario)
print(f"  {binario} en decimal es: {resultado}")

print("\n" + "-" * 50 + "\n")
```

Escribir un programa que pida al usuario un número entero y muestre por pantalla un triángulo rectángulo como el de número de primos.

<b>Salida deseada:</b><br><br>
Introduce el límite del triángulo de primos (entero positivo):  20<br>

1 <br>
3 1 <br>
5 3 1 <br>
7 5 3 1 <br>
11 7 5 3 1 <br>
13 11 7 5 3 1 <br>
17 13 11 7 5 3 1 <br>
19 17 13 11 7 5 3 1

```python
# ========== VERSION MEJORADA ==========
def es_primo(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    
    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True


def obtener_primos_hasta(limite):
    return [num for num in range(2, limite + 1) if es_primo(num)]


# Programa principal
header("TRIANGULO DE NUMEROS PRIMOS")

while True:
    try:
        limite = int(input("Introduce el limite (entero positivo): "))
        if limite > 0:
            break
        print("\t Ingrese un numero positivo")
    except ValueError:
        print("\t Ingrese un numero entero valido")

primos = obtener_primos_hasta(limite)

if not primos:
    print("\t No hay numeros primos en ese rango")
else:
    for i in range(len(primos)):
        # Crear fila con primos desde i hasta 0
        fila = " ".join(str(primos[j]) for j in range(i, -1, -1))
        print(fila)
    print("-" * 50)
```

### Ejercicio 8:
Implemente un programa que sea capaz de calcular el resultado de aplicarla formula siguiente (n i)=n!/(i!\*(n-i)!). Para ello usted debe leer un numero por teclado que debe ser mayor e igual que cero teniendo en cuenta además que n tiene que ser necesariamente mayor que i y además calcular el factorial de un número.

```python
# ========== FUNCION PARA CALCULAR FACTORIAL ==========
def factorial(numero):
    """
    Calcula el factorial de un numero
    factorial(n) = n * (n-1) * (n-2) * ... * 1
    """
    if numero < 0:
        return None

    if numero == 0 or numero == 1:
        return 1

    resultado = 1
    for i in range(2, numero + 1):
        resultado *= i

    return resultado


# ========== FUNCION PARA CALCULAR COEFICIENTE BINOMIAL ==========
def coeficiente_binomial(n, i):
    """
    Calcula C(n,i) = n! / (i! * (n-i)!)

    Condiciones:
    - n >= 0
    - i >= 0
    - n >= i
    """
    # Validar condiciones
    if n < 0 or i < 0:
        return None

    if n < i:
        return None

    # Calcular usando la formula
    numerador = factorial(n)
    denominador = factorial(i) * factorial(n - i)

    if numerador is None or denominador is None:
        return None

    return numerador // denominador


# ========== PROGRAMA PRINCIPAL ==========
header("\t CALCULO DE COEFICIENTE BINOMIAL")
print("\t Formula: C(n,i) = n! / (i! * (n-i)!)")
print("-" * 50)

# Leer valor de n
while True:
    try:
        n = int(input("\t Ingrese n (n >= 0): "))

        if n >= 0:
            break
        else:
            print("\t Error: n debe ser mayor o igual a 0")
    except ValueError:
        print("\t Error: Ingrese un numero entero valido")

# Leer valor de i
while True:
    try:
        i = int(input("\t Ingrese i (i >= 0 y n >= i): "))

        if i >= 0:
            if n >= i:
                break
            else:
                print(f"\t Error: i ({i}) no puede ser mayor que n ({n})")
        else:
            print("\t Error: i debe ser mayor o igual a 0")
    except ValueError:
        print("\t Error: Ingrese un numero entero valido")

# Calcular el coeficiente binomial
resultado = coeficiente_binomial(n, i)

# Mostrar resultado

if resultado is None:
    print("\t Error en el calculo. Verifique los valores ingresados")
else:
    print(f"\t C({n},{i}) = {resultado}")
    print("-" * 50)
    print(f"\t {n}! / ({i}! * ({n}-{i})!) = {resultado}")
    print(f"\t {factorial(n)} / ({factorial(i)} * {factorial(n-i)}) = {resultado}")

print("-" * 50 + "\n")
```

### Ejercicio 9:
Escribir un programa que reciba una cadena de caracteres y devuelva un diccionario con cada palabra que contiene y su frecuencia. Escribir otra función que reciba el diccionario generado con la función anterior y devuelva una tupla con la palabra más repetida y su frecuencia.

```python
# Lorem ipsum dolor sit amet consectetur adipisicing elit. Perferendis voluptatem sit quae. Iste quisquam, libero autem quod, suscipit dolorum blanditiis minima ratione ipsa cum natus facilis, excepturi tempore architecto illo.

# # ========== FUNCION 1: CONTAR FRECUENCIA DE PALABRAS ==========
def contar_frecuencia(texto):
    """
    Recibe una cadena de texto y devuelve un diccionario
    con cada palabra y su frecuencia
    """
    # Limpiar y preparar el texto
    # Convertir a minusculas y eliminar signos de puntuacion
    texto_limpio = ""
    for caracter in texto:
        if caracter.isalnum() or caracter == " ":
            texto_limpio += caracter
        else:
            texto_limpio += " "

    # Dividir en palabras
    palabras = texto_limpio.lower().split()

    # Contar frecuencia
    frecuencia = {}
    for palabra in palabras:
        if palabra in frecuencia:
            frecuencia[palabra] += 1
        else:
            frecuencia[palabra] = 1

    return frecuencia


# ========== FUNCION 2: ENCONTRAR PALABRA MAS REPETIDA ==========
def palabra_mas_repetida(frecuencia):
    """
    Recibe un diccionario de frecuencias y devuelve una tupla
    con la palabra mas repetida y su frecuencia
    """
    if not frecuencia:
        return ("", 0)

    palabra_max = ""
    frecuencia_max = 0

    for palabra, cantidad in frecuencia.items():
        if cantidad > frecuencia_max:
            frecuencia_max = cantidad
            palabra_max = palabra

    return (palabra_max, frecuencia_max)


# ========== PROGRAMA PRINCIPAL ==========
header("ANALISIS DE FRECUENCIA DE PALABRAS")

# Solicitar texto al usuario
texto = input("Ingrese un texto: ")
#texto = "Lorem ipsum dolor sit amet consectetur adipisicing elit. Perferendis voluptatem sit quae. Iste quisquam, libero autem quod, suscipit dolorum blanditiis minima ratione ipsa cum natus facilis, excepturi tempore architecto illo."

# Generar diccionario de frecuencias
frecuencias = contar_frecuencia(texto)

# Mostrar diccionario de frecuencias
print("\n[ FRECUENCIA DE PALABRAS ]")
if frecuencias:
    for palabra, cantidad in sorted(frecuencias.items()):
        print(f"  {palabra}: {cantidad} vez/veces")
else:
    print("  No se encontraron palabras")

print("-" * 50)

# Encontrar palabra mas repetida
palabra, cantidad = palabra_mas_repetida(frecuencias)

# Mostrar resultado
print("\n[ PALABRA MAS REPETIDA ]")
if palabra:
    print(f"  Palabra: '{palabra}'")
    print(f"  Frecuencia: {cantidad} vez/veces")
else:
    print("  No hay palabras para analizar")

print("-" * 50 + "\n")
```
