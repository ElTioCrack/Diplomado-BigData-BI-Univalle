## Creaciﺣﺏn de una base un poco mﺣ۰s compleja

**Nﺣﭦmeros triangulares**

Incluso podemos crear mﺣ۰s series de datos, aprovechemos los bucles para crear mﺣ۰s, quizas la serie de los primeros 100 nﺣﭦmeros triangulares.

> *Un nﺣﭦmero triangular corresponde a la cantidad de puntos que componen a un triﺣ۰ngulo equilatero. Segﺣﭦn el siguiente grﺣ۰fico es facil ver que se obtiene a partir de la suma de los nﺣﭦmeros naturales anteriores a un valor $n$ dado.
![Triangularnumbers](https://study.com/cimages/multimages/16/374px-first_six_triangular_numbers.svg.png)

Usando `for` es fﺣ۰cil encontrar estos nﺣﭦmeros pues $T_n=1+2+3+\cdots+(n-1)+n$, luego el siguiente cﺣﺏdigo permitira ubicar los nﺣﭦmeros triangulares en una lista:


````python
T=[] #Creamos una lista vacﺣ­a
t=0 #Este valor guardarﺣ۰ temporalmente el nﺣﭦmero triangular
for i in range(101): #Observe que range(100) es la lista de valores [0,1,...,99], el comando for toma elementos de una lista.
    t=t+i
    T.append(t)
print(T)
````

````python
# Mostrar solo algunos datos de T
T[:15] 
````

````python
F=[1] #Creamos una lista vacﺣ­a
f=1 #Este valor guardarﺣ۰ temporalmente el nﺣﭦmero factorial
for i in range(1,101): #Observe que range(100) es la lista de valores [0,1,...,99], el comando for toma elementos de una lista.
    f=f*i
    F.append(f)
````

````python
Fib=[1,1] #Creamos una lista vacﺣ­a
for i in range(2,101):
    t=Fib[i-1]+Fib[i-2]
    Fib.append(t)
print(Fib)
````

````python
BaseNumeros=pd.DataFrame({'Triangulares':T,
                         'Fibonacci':Fib,
                         'Constante':22,
                         'Factorial':F})
````

````python
BaseNumeros
````

````python
BaseNumeros.dtypes
````

````python
BaseNumeros.describe()
````

````python
BaseNumeros.Fibonacci=np.array([float(i) for i in BaseNumeros.Fibonacci])
````

````python
BaseNumeros.Fibonacci
````

````python
BaseNumeros.Factorial=pd.to_numeric(BaseNumeros.Factorial)
````

````python
BaseNumeros.dtypes
````

````python
BaseNumeros.describe()
````

## Ejemplo Juguete

Supongamos que podemos medir la velocidad de ciertos animales que vemos pasar en el aire. vemos dos ﺣ۰guilas y dos loros a diferentes velocidad que registramos en un papel.

| Animal | Max Speed |
| -------| --------- |
| Falcon |   380     |
| Falcon |   370     |
| Parrot | 24  |
| Parrot | 26  |

Si queremos colocar dichos datos en un **DataFrame** de pandas, el procedimiento es el siguiente:

````python
import pandas as pd

df = pd.DataFrame({'Animal': ['Falcon', 'Parrot','Falcon', 'Parrot'],
                'Max Speed': [380., 24., 370., 26.]})

df
#print(df) # Look for the differences in the outputs
````

Queremos agrupar los datos de acuerdo a la mediciﺣﺏn hecha:

**Ejemplo:**

````python
df.groupby(['Animal'])
````

Como podemos ver (o mﺣ۰s bien, no podemos), **groupby** no genera un DataFame como tal, sino un objeto **DataFrameGroupBy**, que contiene informaciﺣﺏn sobre los grupos, pero nada mﺣ۰s.

Esto quiere decir, que si queremos informaciﺣﺏn detallada, debemos aplicar alguna operaciﺣﺏn a dicha agrupaciﺣﺏn:

````python
df_proc=df.groupby(['Animal']).mean()
df_proc
````
