# Archivos de Texto

Frecuentemente es deseable que un programa procese o genere archivos de texto.

Generalmente, precisamos procesar archivos con una estructura fija, por ejemplo, un archivo con campos de datos separados por tabulaciones, comas o algún otro delimitador. Cada línea de un archivo de texto estructurado representa datos relacionados a un objeto (datos de una persona, por ejemplo).

Estos archivos pueden ser generados y abiertos por un editor de texto o hasta por un software de hoja de cálculo como Excel. La primera linea, opcionalmente, contiene un encabezado con una descripción de los campos.

Existe dos caracteres comunes usados para delimitar campos, el (\t) o tab y la coma (,). Los separadores de linea, en Linux son caracteres de line feed (\n) que NO son visibles en los editores de texto. En Windows, el terminado de línea es compuesto de dos caractéres consecutivos: el carriage return (\r) y la linea de feed (\n).

Archivos que usan tabulaciones para separar los campos generalmente son guardados con extensión tsv (tab separated velues) y archivos separados por coma son guardados con extensión csv (comma separated values). Sin embargo, esta convención ni siempre es seguida.

### Observación:

No existe un padrón para dar nombres o extensiones a archivos de texto. Usar csv o tsv es apenas una buena práctica para intercambio de archivos separados por tabs o comas. Sin embargo, usted puede usar cualquier extensión u otra extensión.

# Abriendo y leyendo archivos

Primeramente es preciso saber el camino absoluto o relativo donde está el archivo de texto. Se usa la función open() para obtener un manipulador de archivos en Python. Este manipulador de archivos debe ser atribuido a una variable. Las operaciones en el archivo son hechas sobre esta variable. Veamos un ejemplo:

````python
file_loc = './../datasets/elderlyHeightWeight.csv'
f_hand = open(file_loc, 'r')

for linea in f_hand:
    print(linea.strip())

#f_hand.close()
````

En en ejemplo anterior, abrimos un pequeño archivo con los datos de algura y peso de un grupo de hombres y mujeres de la 3ra edad.

La función open() devuelve un objeto archivo, y se usa normalmente con dos argumentos: open(nombre_de_archivo, modo).

El primer argumento es una cadena conteniendo el nombre de archivo. El segundo argumento es otra cadena conteniendo unos pocos caracteres que describen la forma en que el archivo será usado. El modo puede ser 'r' cuando el archivo será solamente leído, 'w' para sólo escribirlo (un archivo existente con el mismo nombre será borrado), y 'a' abre el archivo para agregarle información; cualquier dato escrito al archivo será automáticamente agregado al final. 'r+' abre el archivo tanto para leerlo como para escribirlo. El argumento modo es opcional; si se omite se asume 'r'.

Si se agrega b al modo el archivo se abre en modo binario: ahora los datos se leen y escriben en forma de objetos bytes. Se debería usar este modo para todos los archivos que no contengan texto.

Es una buena práctica usar la declaración <b>with</b> cuando manejamos objetos archivo. Tiene la ventaja que el archivo es cerrado apropiadamente luego de que el bloque termina, incluso si se generó una excepción. También es mucho más corto que escribir los equivalentes bloques try-finally

````python
with open(file_loc) as f:
    datos_leidos = f.read()
    print(f)
f.closed
````

Si no estuvieses usando el bloque with, entonces deberías llamar f.close() para cerrar el archivo e inmediatamente liberar cualquier recurso del sistema usado por este. Si no cierras explícitamente el archivo, el «garbage collector» de Python eventualmente destruirá el objeto y cerrará el archivo por vos, pero el archivo puede estar abierto por un tiempo. Otro riesgo es que diferentes implementaciones de Python harán esta limpieza en diferentes momentos.

## 6.1.1 Abrir Automáticamente en el Navegador

Si el fichero que se ha creado con Python es un página web, puede ser de utilidad que el fichero se abra automáticamente en el navegador para ver el resultado inmediatamente.

Para ello se puede utilizar la función open() del módulo webbrowser de la biblioteca estándar, que abre el fichero indicado con la aplicación asociada en el sistema operativo.

````python
import os
import webbrowser

carpeta = "3.1.2"
os.makedirs(carpeta, exist_ok=True)

ruta = os.path.join(carpeta, "prueba.html")

with open(ruta, mode="w", encoding="utf-8") as fichero:
    print("<!DOCTYPE html>", file=fichero)
    print('<html lang="es">', file=fichero)
    print("<head>", file=fichero)
    print('  <meta charset="utf-8">', file=fichero)
    print("  <title>HTML 5</title>", file=fichero)
    print('  <meta name="viewport" content="width=device-width, initial-scale=1.0">', file=fichero)
    print("</head>", file=fichero)
    print("", file=fichero)
    print("<body>", file=fichero)
    print("  <p>Esta página es una página HTML 5 válida.</p>", file=fichero)
    print("</body>", file=fichero)
    print("</html>", file=fichero)

webbrowser.open(ruta)
````

Es fundamental que el juego de caracteres en el que se guarda el fichero (línea 5 en el ejemplo anterior) sea el mismo que el que se declara en la etiqueta <meta> (línea 9 en el ejemplo anterior). Si no es así, el navegador tendrá problemas al mostrar caracteres "no ingleses" (acentos, eñes, etc.). En el ejemplo anterior se utiliza UTF-8, el juego de caracteres recomendado para páginas web.

### Ejercicio 1:
Escribir un programa, llamado head que reciba un archivo y un número N e imprima las primeras N líneas del archivo.

````python
# ========== EJERCICIO 1: HEAD.PY ==========
# import os

def head(archivo, N):
    if not os.path.exists(archivo):
        return print("Archivo no encontrado.")

    with open(archivo, 'r') as f:
        for i, linea in enumerate(f):
            if i > N: # N líneas de datos + 1 de cabecera
                break
            print(linea.strip())

archivo = './../datasets/elderlyHeightWeight.csv'
try:
    N = int(input("Lineas extras: "))
    head(archivo, N)
except:
    print("Error en el número.")
````

### Ejercicio 2:
Escribir un programa, llamado cp.py, que copie todo el contenido de un archivo (sea de texto o binario) a otro, de modo que quede exactamente igual. Nota: utilizar archivo.read(bytes) para leer como máximo una cantidad de bytes.

````python
# ========== EJERCICIO 2: CP.PY ==========
import os

def cp(origen, destino):
    if not os.path.exists(origen):
        return print("Error: El origen no existe.")

    # Crear carpeta destino si no existe
    os.makedirs(os.path.dirname(destino), exist_ok=True)
    bloque = 4096
    with open(origen, 'rb') as f_origen:
        with open(destino, 'wb') as f_destino:
            while True:
                datos = f_origen.read(bloque)
                if not datos:
                    break
                f_destino.write(datos)

    print(f"Copiado a: {destino}")

# --- EJECUCION ---
archivo_origen = './../datasets/elderlyHeightWeight.csv'
archivo_destino = './3.1.2/copia_elderlyHeightWeight.csv'

cp(archivo_origen, archivo_destino)
````

### Ejercicio 3:
Escribir un programa, llamado wc.py que reciba un archivo, lo procese e imprima por pantalla cuántas líneas, cuantas palabras y cuántos caracteres contiene el archivo.

````python
# ========== EJERCICIO 3: WC.PY ==========
def wc(archivo):
    lineas = 0
    palabras = 0
    caracteres = 0

    with open(archivo, 'r', encoding='utf-8') as f:
        for linea in f:
            lineas += 1
            caracteres += len(linea)
            palabras += len(linea.split())

    print(f"Archivo: {archivo}")
    print("-" * 40)
    print(f"  Lineas:     {lineas}")
    print(f"  Palabras:   {palabras}")
    print(f"  Caracteres: {caracteres}")
    print("-" * 40)

    return lineas, palabras, caracteres

# --- EJECUCION ---
archivo = './../datasets/elderlyHeightWeight.csv'
wc(archivo)
````

# 6.2 El Módulo ```csv``` 

Aunque podemos leer y escribir datos de/para un archivo usando Python puro, existen bibliotecas que facilitan estas operaciones para archivos separados por delimitadores. La biblioteca más utilizada es [```csv```](http://www.pythonforbeginners.com/systems-programming/using-the-csv-module-in-python/). A pesar del nombre csv, maneja archivos separados por cualquier delimitador que especifiquemos. Para usarlo, necesitamos hacer una ```import```.

````python
import csv
````

Para leer datos de un archivo con separadores se usa la función ```reader()```. La función ```reader()``` copia cada línea del archivo y crea un objeto de clase ```reader```.

Un método de ```reader```es un ```.next()```que usaremos para acceder una linea por vez. Una vez que procesamos una línea, el objeto .next() apuntará para la próxima línea.

### Observación:

De aqui en adelante, vamos a usar las declaraciones `with/as` para manipular archivos. Esta es la forma mas usada a partir de Python3 y deja el código más legible.

````python
import csv

with open('./../datasets/elderlyHeightWeight.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter='\t') # Observe que estamos indicando que el delimitador es un tab
    header = next(reader)  # Retorna una lista de strings
    print (header)  # Observe que next(reader) retorna una lista de strings
    print () # Imprime una línea en blanco

    for i in range(4):
        print (next(reader)) # Imprime las 4 líneas siguientes después del encabezado.
````

Obsere que todos los valores son leidos como strings. En caso de usar los valores en otras operaciones, tendrá que convertirlos en int o float.

Podemos usar el módulo csv para seleccionar columnas enteras de la siguiente forma.

````python

with open('./../datasets/elderlyHeightWeight.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter='\t') # delimitador es coma

    # use el método next() para copiar al primera linea con el encabezado
    headers = next(reader)  #headers ahora es una lsita de strings
    print(headers)
    
    # weight es el tercer campo (índice 2), height es el cuarto (índice 3)
    
    weight  = ['Weight'] # lista para guardar los weights, inicializando con el encabezado
    height = ['Height']  # lista para guardar los heights, inicializando con el encabezado

    for row in reader:  # Implicitamente es llamado el método next para acceder a todas las lineas del archivo
        weight.append(row[2])  # Adiciona los weight en la lista
        height.append(row[3])  #Adiciona los height en la lista
    
    print (weight)
    print (height)
````

Use el módulo csv para abrir el archivo ```elderlyHeightWeight.csv```. Escriba un nuevo archivo (male_data.tsv) conteniendo apenas los datos de los hombres (Gender M).

Obs: ```csv.reader``` son listas. El método ```.join()``` concatena elementos de listas usando una string como delimitador.

````python
import csv

#Abrindo dos archivos de una vez
with open('./../datasets/elderlyHeightWeight.csv', 'r') as csvfile, open('./3.1.2/male_data.tsv', 'w') as new_file:
    reader = csv.reader(csvfile, delimiter='\t') # define the field delimiter
    # Use el método next() para copiar la primera linea con los encabezados
    headers = next(reader)
    new_file.write('\t'.join(headers) + '\n')
    for row in reader:  
        if row[0]=="M":
            new_file.write('\t'.join(row) + '\n')


print("Archivo male_data.tsv creado con exito")
````

### 6.2.2 Adicionando datos a un archivo ya existente

USe el argumento ```'a'``` de append en la función ```open()```. De esta forma el contenido existente del archivo será preservado y las nuevas líneas serán escritas al final del archivo. Vamos a abrir el archivo con los dados de los hombres que acabamos de generar en modo append y adicionar los datos de las mujeres. Observe aquí una nueva estructura, o try/except, usada en esta sección para imprimir un mensaje de error caso algún archivo no exista al intentar abrirlo.

````python
import csv

try:
    with open('./3.1.2/male_data.tsv', 'a') as new_file, open('./../datasets/elderlyHeightWeight.csv', 'r') as f_hand:
        reader = csv.reader(f_hand, delimiter='\t')
        female_data = [line for line in reader if line[3] == 'Female']  # Cambiar line[0] a line[3] y 'F' a 'Female'
        for line in female_data:
            new_file.write('\t'.join(line) + '\n')
    print("Datos de mujeres agregados a male_data.tsv")
except FileNotFoundError:
    print('The file does not exist.')
````

### Ejercicio 4:

Procese el archivo para responder:

1) Número de individuos de cada sexo

2) Média de peso y de edad por sexo

````python
# ============================================
# EJERCICIO 4 - 3.2.2
# ============================================
# Lic. Ilian Joseph Felipez Vaca
# ============================================

import csv

archivo = './../datasets/elderlyHeightWeight.csv'

datos = {
    'M': {'count': 0, 'sum_edad': 0, 'sum_peso': 0, 'genero_nombre': 'Male'},
    'F': {'count': 0, 'sum_edad': 0, 'sum_peso': 0, 'genero_nombre': 'Female'}
}

with open(archivo, 'r') as f:
    reader = csv.reader(f, delimiter='\t')
    next(reader)  # saltar cabecera
    
    for row in reader:
        if len(row) < 3:
            continue  # saltar lineas mal formadas
            
        genero = row[0].strip()
        edad = int(row[1])
        peso = float(row[2])
        
        if genero in datos:
            datos[genero]['count'] += 1
            datos[genero]['sum_edad'] += edad
            datos[genero]['sum_peso'] += peso

# ========== MOSTRAR RESULTADOS ==========
print("\n" + "=" * 50)
print("RESULTADOS DEL ANALISIS")
print("=" * 50)

print("\n1) NUMERO DE INDIVIDUOS POR SEXO:")
for genero in datos:
    print(f"\t{datos[genero]['genero_nombre']}: {datos[genero]['count']}")

print("\n2) MEDIA DE EDAD Y PESO POR SEXO:")
print("\t" + "-" * 35)
for genero in datos:
    count = datos[genero]['count']
    if count > 0:
        media_edad = datos[genero]['sum_edad'] / count
        media_peso = datos[genero]['sum_peso'] / count
        print(f"\t{datos[genero]['genero_nombre'].upper()}:")
        print(f"\t  Edad promedio: {media_edad:.1f} anos")
        print(f"\t  Peso promedio: {media_peso:.1f} kg")
        print("\t" + "-" * 35)
````

### Ejercicio 6:
Escribir un programa, llamado rot13.py que reciba un archivo de texto de origen y uno de destino, de modo que para cada línea del archivo origen, se guarde una línea cifrada en el archivo destino. El algoritmo de cifrado a utilizar será muy sencillo: a cada caracter comprendido entre la a y la z, se le suma 13 y luego se aplica el módulo 26, para obtener un nuevo caracter.

Nota: revisar https://uniwebsidad.com/libros/algoritmos-python/capitulo-11/persistencia-de-datos

````python
# Escriba el código aqui
# ========== EJERCICIO 6: ROT13.PY ==========
import os

def rot13_caracter(caracter):
    """
    Cifra un caracter usando ROT13
    Solo afecta letras de la a a la z (minusculas) y A a Z (mayusculas)
    """
    if 'a' <= caracter <= 'z':
        offset = ord('a')
        return chr(((ord(caracter) - offset + 13) % 26) + offset)
    elif 'A' <= caracter <= 'Z':
        offset = ord('A')
        return chr(((ord(caracter) - offset + 13) % 26) + offset)
    else:
        return caracter


def rot13_linea(linea):
    """
    Cifra una linea completa usando ROT13
    """
    resultado = ""
    for caracter in linea:
        resultado += rot13_caracter(caracter)
    return resultado


def rot13_archivo(origen, destino):
    """
    Lee un archivo origen, cifra cada linea con ROT13
    y guarda el resultado en el archivo destino
    """
    try:
        with open(origen, 'r', encoding='utf-8') as f_origen:
            with open(destino, 'w', encoding='utf-8') as f_destino:
                
                for linea in f_origen:
                    linea_cifrada = rot13_linea(linea.rstrip('\n'))
                    f_destino.write(linea_cifrada + '\n')
        
        print(f"✓ Archivo cifrado exitosamente")
        print(f"  Origen:  {origen}")
        print(f"  Destino: {destino}")
        
    except FileNotFoundError:
        print(f"Error: No se encuentra el archivo '{origen}'")
    except Exception as e:
        print(f"Error inesperado: {e}")


# ========== EJECUCION ==========
if __name__ == "__main__":
    # Crear carpeta 3.1.2 si no existe
    carpeta = "3.1.2"
    if not os.path.exists(carpeta):
        os.makedirs(carpeta)
        print(f"Carpeta '{carpeta}' creada")
    
    # Archivos dentro de la carpeta 3.1.2
    archivo_origen = os.path.join(carpeta, 'mensaje.txt')
    archivo_destino = os.path.join(carpeta, 'mensaje_cifrado.txt')
    
    # Crear archivo de ejemplo si no existe
    if not os.path.exists(archivo_origen):
        with open(archivo_origen, 'w', encoding='utf-8') as f:
            f.write("Hola mundo\n")
            f.write("Este es un mensaje secreto\n")
            f.write("Python es genial\n")
        print(f"Archivo de ejemplo creado: {archivo_origen}")
    
    # Cifrar el archivo
    rot13_archivo(archivo_origen, archivo_destino)
    
    # Mostrar resultado
    print("\n--- CONTENIDO ORIGINAL ---")
    with open(archivo_origen, 'r', encoding='utf-8') as f:
        print(f.read())
    
    print("--- CONTENIDO CIFRADO (ROT13) ---")
    with open(archivo_destino, 'r', encoding='utf-8') as f:
        print(f.read())
    
    # Demostrar que ROT13 es reversible
    archivo_descifrado = os.path.join(carpeta, 'mensaje_descifrado.txt')
    print("--- CONTENIDO DESCIFRADO (ROT13 nuevamente) ---")
    rot13_archivo(archivo_destino, archivo_descifrado)
    with open(archivo_descifrado, 'r', encoding='utf-8') as f:
        print(f.read())
````

````python
# Escriba el código aqui
# ========== EJERCICIO 7: GUARDAR DICCIONARIO EN ARCHIVO ==========
import os

def guardar_datos(diccionario, nombre_archivo):
    """
    Guarda un diccionario en un archivo con formato clave,valor
    Cada linea del archivo contiene: clave,valor
    """
    try:
        with open(nombre_archivo, 'w', encoding='utf-8') as f:
            for clave, valor in diccionario.items():
                f.write(f"{clave},{valor}\n")
        
        print(f"✓ Diccionario guardado en: {nombre_archivo}")
        return True
        
    except Exception as e:
        print(f"Error al guardar: {e}")
        return False


def cargar_datos(nombre_archivo):
    """
    Carga un diccionario desde un archivo con formato clave,valor
    """
    diccionario = {}
    
    try:
        with open(nombre_archivo, 'r', encoding='utf-8') as f:
            for linea in f:
                linea = linea.strip()
                if linea:
                    clave, valor = linea.split(',', 1)
                    diccionario[clave] = valor
        
        print(f"✓ Diccionario cargado desde: {nombre_archivo}")
        return diccionario
        
    except FileNotFoundError:
        print(f"Error: No se encuentra el archivo '{nombre_archivo}'")
        return {}
    except Exception as e:
        print(f"Error al cargar: {e}")
        return {}


# ========== EJECUCION ==========
if __name__ == "__main__":
    # Crear carpeta 3.1.2 si no existe
    carpeta = "3.1.2"
    if not os.path.exists(carpeta):
        os.makedirs(carpeta)
    
    # Diccionario de ejemplo
    datos_ejemplo = {
        "nombre": "Ana",
        "edad": "25",
        "ciudad": "Barcelona",
        "profesion": "ingeniera"
    }
    
    # Guardar en archivo dentro de la carpeta
    archivo = os.path.join(carpeta, "datos.txt")
    guardar_datos(datos_ejemplo, archivo)
    
    # Cargar y verificar
    datos_cargados = cargar_datos(archivo)
    print(f"\nDiccionario cargado: {datos_cargados}")
    
    # Mostrar contenido del archivo
    print("\n--- CONTENIDO DEL ARCHIVO ---")
    with open(archivo, 'r', encoding='utf-8') as f:
        print(f.read())
````

## 6.3.1 Errores de sintaxis¶

Los errores de sintaxis, también conocidos como errores de interpretación, son quizás el tipo de queja más común que se tiene cuando todavía estás aprendiendo Python:

````python
while True print('Hola mundo')
````

El intérprete repite la línea culpable y muestra una pequeña ‘flecha’ que apunta al primer lugar donde se detectó el error. Este es causado por (o al menos detectado en) el símbolo que precede a la flecha: en el ejemplo, el error se detecta en la función print(), ya que faltan dos puntos (':') antes del mismo. Se muestran el nombre del archivo y el número de línea para que sepas dónde mirar en caso de que la entrada venga de un programa.

## 6.3.2 Excepciones

Incluso si la declaración o expresión es sintácticamente correcta, puede generar un error cuando se intenta ejecutarla. Los errores detectados durante la ejecución se llaman excepciones, y no son incondicionalmente fatales. Sin embargo, la mayoría de las excepciones no son manejadas por los programas, y resultan en mensajes de error como los mostrados aquí:

````python
10 * (1/0)
````

Las excepciones vienen de distintos tipos, y el tipo se imprime como parte del mensaje: los tipos en el ejemplo son: <b>ZeroDivisionError, NameError y TypeError</b>. La cadena mostrada como tipo de la excepción es el nombre de la excepción predefinida que ocurrió. Esto es verdad para todas las excepciones predefinidas del intérprete, pero no necesita ser verdad para excepciones definidas por el usuario (aunque es una convención útil). Los nombres de las excepciones estándar son identificadores incorporados al intérprete (no son palabras clave reservadas).

## 6.3.3 Manejando excepciones

Es posible escribir programas que manejen determinadas excepciones. La declaración try funciona de la siguiente manera:

<ul>
    <li>Primero, se ejecuta el bloque try.</li>
    <li>Si no ocurre ninguna excepción, el bloque except se salta y termina la ejecución de la declaración try.</li>
    <li>Si ocurre una excepción durante la ejecución del bloque try, el resto del bloque se salta. Luego, si su tipo coincide con la excepción nombrada luego de la palabra reservada except, se ejecuta el bloque except, y la ejecución continúa luego de la declaración try.</li>
    <li>Si ocurre una excepción que no coincide con la excepción nombrada en el except, esta se pasa a declaraciones try de más afuera; si no se encuentra nada que la maneje, es una excepción no manejada, y la ejecución se frena con un mensaje como los mostrados arriba.</li>
</ul>
