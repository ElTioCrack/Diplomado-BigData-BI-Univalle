## Introducción

El paquete para construir gráficos por excelencia en Python es [Matplotlib](https://matplotlib.org/3.2.2/index.html), una de las librerías más usadas para crear todo tipo de visualizaciones  . Para nosotros es muy importante manejar esta herramienta ya que en múltiples ocasiones la dificultad de muchas explicaciones se resuelven fácilmente con un buen gráfico. En este cuaderno veremos algunos componentes básicos de la graficación con Matplotlib, las partes del gráfico, la definición de `figure()`, unas configuraciones simples y algunas personalizaciones elementales. 

## Instalando MatplotLib en Linux y Windows

Para instalar Matplotlib en Windows y Ubuntu, puedes seguir los siguientes pasos:

**Windows:**

1. Abre el menú "Inicio" y busca "Command Prompt" (Símbolo del sistema). Haz clic derecho sobre él y selecciona "Ejecutar como administrador" para abrir una ventana de línea de comandos con privilegios de administrador.

2. En la ventana de línea de comandos, ingresa el siguiente comando para instalar Matplotlib utilizando pip, que es el sistema de gestión de paquetes de Python:
   
   ```bash
   pip install matplotlib
   ```

3. Espera a que la instalación se complete. Una vez finalizada, Matplotlib estará instalado en tu sistema.

**Ubuntu:**

1. Abre una terminal. Puedes hacerlo presionando `Ctrl + Alt + T` o buscando "Terminal" en el menú de aplicaciones.

2. Actualiza la lista de paquetes utilizando el siguiente comando:

   ```bash
   sudo apt update
   ```

3. Instala el paquete `python3-matplotlib` utilizando el siguiente comando:

   ```bash
   sudo apt install python3-matplotlib
   ```

4. Confirma la instalación ingresando `Y` si se te solicita y espera a que se complete.

Una vez que hayas seguido estos pasos, Matplotlib estará instalado en tu sistema. Ahora puedes importarlo en tus scripts de Python y comenzar a crear visualizaciones gráficas. Por ejemplo:

```python
import matplotlib.pyplot as plt

# Tu código de visualización aquí

plt.show()  # No muestra nada, pero tampoco debe generar un error
```

En Windows, asegúrate de que Python y pip estén correctamente configurados en tus variables de entorno para que puedas acceder a ellos desde cualquier ubicación en la línea de comandos.

## ¿Cómo funciona?

Matplotlib grafica sus datos sobre una instancia llamada `Figure`, cada una de estas contendrá uno o más `Axes` (un área en la que los puntos tienen coordenadas). Usualmente se utiliza la interfaz [pyplot](https://matplotlib.org/3.7.2/api/pyplot_summary.html) basada en algunos estilos y comandos de MATLAB. 

Veamos un sencillo ejemplo:

````python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
fig, ax = plt.subplots()  # Creamos una figura con un sistema de ejes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Pintamos algunos datos.
````

Sin necesidad de fijar los ejes de una figura podemos hacer el mismo gráfico:

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 5)
y = np.random.randint(1, 10, size=4)

plt.plot(x, y)
plt.show()
````

No obstante, recordemos que podemos usar instancias para expresar tanto a la figura como los ejes sobre la figura:

````python
# plt.figure() Crea una nueva figura vacía.
# Una figura es el espacio en blanco donde se pueden colocar uno o varios gráficos
fig = plt.figure()

# Agrega un subplot (subgráfico) a la figura
# Un subplot es un área en la figura donde se pueden colocar gráficos
# El argumento 111 significa que se está creando un único subplot en una cuadrícula de 1 fila y 1 columna
ax = fig.add_subplot(111)

# Traza una línea que conecta los puntos (1, 10), ... en el subplot
# La línea se traza en color azul claro (lightblue) con un grosor de línea de 3
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)

# Agrega puntos dispersos en el subplot en las posiciones (0.3, 11), ...
# Los puntos se representan en color verde oscuro (darkgreen) y 
# se utilizan marcadores en forma de triángulos (^)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')

ax.set_xlim(0, 4.5) # Establece los límites del eje x en el subplot (de 0 a 4.5)

# Muestra la figura completa con el gráfico y los subgráficos creados
plt.show()
````

En este ejemplo utilizamos `plt.figure()` para definir la figura y `fig.add_subplot(111)` para crear la *'zona de dibujo'*. Definamos sobre una misma figura varias zonas de dibujo:

````python
import matplotlib.pyplot as plt

# Crear una nueva figura vacía
fig = plt.figure()

# Subplot 1: Primer subgráfico en una cuadrícula de 2x2
ax = fig.add_subplot(221)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3, label="línea")
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^', label="marker")
ax.legend()  # Mostrar leyenda
plt.title("Primera 'zona de dibujo'")  # Agregar título

# Subplot 2: Segundo subgráfico en una cuadrícula de 2x2
ax2 = fig.add_subplot(222)
ax2.plot(x, x, label='línea')
ax2.legend()  # Mostrar leyenda
plt.title("Segunda 'zona de dibujo'")  # Agregar título

# Subplot 3: Tercer subgráfico en una cuadrícula de 2x2
ax3 = fig.add_subplot(223)
ax3.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3, label="línea")
ax3.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^', label="marker")
ax3.legend()  # Mostrar leyenda
plt.title("Tercera 'zona de dibujo'")  # Agregar título

# Subplot 4: Cuarto subgráfico en una cuadrícula de 2x2
ax4 = fig.add_subplot(224)
ax4.plot(x, x, label='línea')
ax4.legend()  # Mostrar leyenda
plt.title("Cuarta 'zona de dibujo'")  # Agregar título

# Mostrar la figura completa con los subgráficos
plt.show()
````

Las zonas de dibujo (`Axes`) siempre deben vincularse a una Figura. Podemos fijar un título para cada zona usando `set_title()`. Por cada `Axes` tenemos 2 o tres ejes (`Axis`), entre sus funcioes útiles tenemos:

**Para controlar el límite de los ejes:**
* `axes.Axes.set_xlim()`
* `axes.Axes.set_ylim()`

**Para fijar una etiqueta en el eje**

* `set_xlabel()`
* `set_ylabel()`


Veamos un ejemplo:

````python
import matplotlib.pyplot as plt

# Crear una nueva figura vacía
fig = plt.figure()

# Subplot 1: Primer subgráfico en una cuadrícula de 1x2
ax = fig.add_subplot(121)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3, label="línea")
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^', label="marker")

# Configuraciones del subgráfico 1
ax.set_xlim(0, 4.5)  # Límites en el eje x
ax.set_ylim(0, 31)   # Límites en el eje y
ax.set_ylabel("Eje y")  # Etiqueta del eje y
ax.set_xlabel("Eje x")  # Etiqueta del eje x
ax.legend()  # Mostrar leyenda
ax.set_title("Primera 'zona de dibujo'")  # Agregar título

# Subplot 2: Segundo subgráfico en una cuadrícula de 1x2
ax2 = fig.add_subplot(122)
ax2.plot(x, x, label='línea')

# Configuraciones del subgráfico 2
ax2.legend()  # Mostrar leyenda
ax2.set_title("Segunda 'zona de dibujo'")  # Agregar título

# Mostrar la figura completa con los subgráficos
plt.show()
````

````python
# Crear un arreglo de valores x en el rango de -10 a 10, con incrementos de 0.1
x = np.arange(-10, 10, 0.1)
# Calcular los valores y correspondientes usando la función seno
y = np.sin(x)
# Graficar los puntos (x, y) utilizando marcadores verdes en forma de estrella (*)
plt.plot(x, y, 'g*')
# Etiqueta para el eje y
plt.ylabel('Un gráfico de sen(x)')
# Mostrar la gráfica
plt.show()
````

**Varias funciones en un mismo gráfico:**

Esta línea de código realiza una gráfica con tres conjuntos de datos en un solo gráfico utilizando la función `plot` de Matplotlib. Aquí hay una descripción detallada de lo que hace cada parte de la línea:

- `plt.plot(x, np.sin(x) + np.cos(x), 'r--', x, np.sin(x), 'bs', x, np.cos(x), 'g^')`:

  - `x`: Los valores en el eje x, que se supone que ya han sido definidos previamente.
  - `np.sin(x) + np.cos(x)`: El conjunto de datos que se obtiene al sumar los valores de la función seno y la función coseno para cada valor de x.
  - `'r--'`: Este es el estilo de línea y color para el primer conjunto de datos, que es la suma de seno y coseno. `'r--'` indica que se debe usar una línea roja discontinua (`--` representa la línea discontinua) para este conjunto de datos.
  - `x, np.sin(x), 'bs'`: Este es el segundo conjunto de datos, que corresponde a la función seno. `'bs'` significa que se deben utilizar cuadrados azules (`b` para azul y `s` para cuadrados) para representar este conjunto de datos.
  - `x, np.cos(x), 'g^'`: Este es el tercer conjunto de datos, que corresponde a la función coseno. `'g^'` significa que se deben utilizar triángulos verdes (`g` para verde y `^` para triángulos) para representar este conjunto de datos.

````python
plt.plot(x,np.sin(x)+np.cos(x) , 'r--', x, np.sin(x), 'bs', x, np.cos(x), 'g^')
plt.show()
````

En resumen, esta línea de código crea un gráfico con tres conjuntos de datos: la suma de las funciones seno y coseno representada por una línea roja discontinua, la función seno representada por cuadrados azules y la función coseno representada por triángulos verdes. Los conjuntos de datos se superponen en un solo gráfico para comparar visualmente sus comportamientos.

````python
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots()  # Creamos una instancia de figura y una zona de dibujo.
ax.plot(x, x, label='Función lineal')  
ax.plot(x, x**2, label='Función cuadrática')  
ax.plot(x, x**3, label='Función cúbica') 
ax.set_xlabel('eje x')  
ax.set_ylabel('eje y')  
ax.set_title("Tres funciones")  
ax.legend()  
````

### Graficador de funciones

````python
def graffun(nf,xmin,xmax,ymin,ymax):
    try:
        x = np.linspace(xmin, xmax, 100)
        y=eval(nf)
        plt.plot(x,y)
        plt.axis([xmin,xmax,ymin,ymax])
        plt.grid(True)
    except:
        print('Error!!')


interact(graffun,nf='(x-3)**2',xmin=0,xmax=2*np.pi,ymin=-1,ymax=1)  
````
